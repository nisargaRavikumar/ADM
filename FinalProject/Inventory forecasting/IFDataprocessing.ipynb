```python
calender = pd.read_csv(r'C:\Users\nisar\Downloads\m5-forecasting-accuracy\calendar.csv')
```

```python
DTdate={}
for index, row in calender.iterrows():
    wk=row['wm_yr_wk']
    d=row['date']
    if wk not in DTdate:
        DTdate[wk]=[d]
    else:
        DTdate[wk].append(d)
```
